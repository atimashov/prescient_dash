```python
X = df[['Press', 'Outdoor', 'Radio', 'TV', 'Programmatic', 'Search',
       'Fbk & Inst', 'YouTube', 'Other Digital']]
```

# Traffic

```python
Y1 = df[['Traffic']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y1_train, Y1_test = train_test_split(X, Y1, test_size = 0.1, random_state = 101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
print([X_train, Y1_train, X_test, Y1_test])
```

```python
print(X_test)
```

```python
# X_dummy = pd.DataFrame(X_test)
# print(X_dummy)
```

```python
# Pass a series in append() to append a row in dataframe  
X_test = X_test.append(pd.Series([0.00, 0.00, 0.00, 0.00, 45055.00, 9988.00, 44302.00, 0.00, 0.00], index=X_test.columns), ignore_index=True)
```

```python
print(X_test)
```

```python
lm = LinearRegression()
```

```python
lm.fit_intercept
```

```python
lm.fit(X_train, Y1_train)
```

```python
print(lm.intercept_)
```

```python
print(lm.coef_)
```

```python
X_train.columns
```

```python
traffic_predictions = lm.predict(X_test)
```

```python
print(traffic_predictions)
```

```python
Y1_test
```

```python
sns.distplot(Y1_test-traffic_predictions)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(Y1_test, traffic_predictions))
print('MSE:', metrics.mean_squared_error(Y1_test, traffic_predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y1_test, traffic_predictions)))
```
